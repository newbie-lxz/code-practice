```python
import requests
import pandas as pd
import time
import random

print("开始爬取")

all_bili_data = []
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36',
    'Referer': 'https://www.bilibili.com/'
}

target_regions = {
    1: "动画", 3: "音乐", 4: "游戏", 5: "娱乐", 36: "科技", 
    188: "数码", 201: "知识", 160: "生活", 211: "美食", 
    217: "动物圈", 223: "汽车", 155: "时尚", 119: "搞笑", 
    181: "影视", 177: "纪录片"
}

for rid, rname in target_regions.items():
    print(f"\n【{rname}】分区：")

    for page in range(1, 21):
        url = f"https://api.bilibili.com/x/web-interface/dynamic/region?ps=20&rid={rid}&pn={page}"
        
        try:
            response = requests.get(url, headers=headers, timeout=10)
            res_json = response.json()
            
            if res_json['code'] == 0:
                archives = res_json['data'].get('archives', [])
                if not archives:
                    break 
                    
                for v in archives:
                    stat = v['stat']
                    all_bili_data.append({
                        '视频标题': v['title'],
                        '所属分区': rname,
                        'UP主': v['owner']['name'],
                        '播放量': stat['view'],
                        '弹幕数': stat['danmaku'],
                        '收藏数': stat['favorite'],
                        '投币数': stat['coin'],
                        '分享数': stat['share'],
                        '点赞数': stat['like']
                    })
            else:
                break
            
            
            time.sleep(random.uniform(0.6, 1.2))
            
        except Exception as e:
            print(f"{rname}分区报错: {e}")
            break

# 最终保存
if all_bili_data:
    df = pd.DataFrame(all_bili_data)
    df = df.drop_duplicates(subset=['视频标题', 'UP主'])
    file_name = "数据_2.xlsx"
    df.to_excel(file_name, index=False)
    print(f"\n结束，共{len(df)} 条。")
else:
    print("error")
```
